# Spatial  clustering

## Kmeans

首先，我先隨便選一個值當作k來觀察結果，這邊的k=10

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import time
from sklearn.cluster import AgglomerativeClustering
import numpy as np

col=['blue','red','#76FF03','orange','#FFEB3B','#29B6F6','#ff8a80','gold','#607D8B','#BA68C8','#18FFFF','#6D7480','#EC407A']
data=pd.read_csv('./station_data.csv')
data=data.iloc[:,[2,3]] #extract latitude and longitude
```

```python
t_start=time.time()
k=10 #set clustering number value
kmeans=KMeans(n_clusters=k,n_init=15).fit(data)
plt.figure(figsize=(7,6),dpi=100)

for idx in range(0,k): #plot different color for each cluster
    lat=[]
    lon=[]
    for index,value in data.iterrows():
        if(kmeans.labels_[index]==idx):
            lat.append(value['end station latitude'])
            lon.append(value['end station longitude'])
    plt.scatter(lat,lon,s=5,color=col[idx],label=idx)

print('It cost '+str(round(time.time()-t_start,6))+' s.')
plt.title('cluster with KMeans k='+str(k))
plt.xlabel('latitude')
plt.ylabel('longitude')
plt.legend()
plt.show()
```

很明顯的可以看到這樣的切法過細，很多相近的點也被切成兩群，因此我用了elbow method觀察每次切割後的距離差，來找k在哪個值時比較適合

```python
cluster_range=range(1,20)
errors=[]
plt.figure(figsize=(7,5))

for k in cluster_range: #record sum of squared distances to their closest center(inertia_) for each k values 
	kmeans=KMeans(n_clusters=k,n_init=15).fit(data)
	errors.append(kmeans.inertia_)

plt.plot(cluster_range,errors,marker='o')
plt.xticks(np.linspace(0,20,21))
plt.title('elbow method')
plt.xlabel('cluster number')
plt.ylabel('sum of squared distances')
plt.show()
```

可以看到在k=4時，差異值下降的速率低了不少，也許可以將k設為4

```python
t_start=time.time()
k=4
kmeans=KMeans(n_clusters=k,n_init=15).fit(data)
plt.figure(figsize=(7,6),dpi=100)

for idx in range(0,k): #plot different color for each cluster
	lat=[]
	lon=[]
	for index,value in data.iterrows():
		if(kmeans.labels_[index]==idx):
			lat.append(value['end station latitude'])
			lon.append(value['end station longitude'])
	plt.scatter(lat,lon,s=5,color=col[idx],label=idx)

print('It cost '+str(round(time.time()-t_start,6))+' s.')
plt.title('cluster with KMeans k='+str(k))
plt.xlabel('latitude')
plt.ylabel('longitude')
plt.legend()
plt.show()
```

結果好了不少，但直覺上應該是右邊一大群，左邊可能一群或兩群，因此試著將k取到13，再用工人智慧合併

```python
t_start=time.time()
k=13
kmeans=KMeans(n_clusters=k,n_init=15).fit(data)
plt.figure(figsize=(7,6),dpi=100)

for idx in range(0,k): #plot different color for each cluster
	lat=[]
	lon=[]
	for index,value in data.iterrows():
		if(kmeans.labels_[index]==idx):
			lat.append(value['end station latitude'])
			lon.append(value['end station longitude'])
	plt.scatter(lat,lon,s=5,color=col[idx],label=idx)

print('It cost '+str(round(time.time()-t_start,6))+' s.')
plt.title('cluster with KMeans k='+str(k))
plt.xlabel('latitude')
plt.ylabel('longitude')
plt.legend()
plt.show()
```

成功將點都分群了出來，再來就是看要應用在什麼樣的地方，決定這些群怎麼合併會比較好，這邊就不再多做

比較訝異的是紅色，編號一的群，竟然跨過了中間明顯的分隔區，將兩邊的點分成了同一群，可能要在切的更細才能解決這種情況

而很有趣的，右上方一顆點自己切成一組，下方距離較遠的點也自成一群，算是解決了outlier影響分群的問題

## DBScan

第二個clustering的方法是DBScan，相較於KMean，他擁有更佳的抵抗outlier的能力。

觀察上圖兩相鄰點的距離，故試試將eps=0.005，min_samples=2，metric為預設的euclidean

```python
from sklearn.cluster import DBSCAN
```

```python
t_start=time.time()
dbscan=DBSCAN(eps=0.005,min_samples=2).fit(data)
plt.figure(figsize=(7,6),dpi=100)
k=len(set(dbscan.labels_))-(1 if -1 in dbscan.labels_ else 0)

for idx in range(-1,k):
	lat=[]
	lon=[]
	for index,value in data.iterrows():
		if(dbscan.labels_[index]==idx):
			lat.append(value['end station latitude'])
			lon.append(value['end station longitude'])
	if idx==-1:
		plt.scatter(lat,lon,s=5,color='black',label='outlier')
	else:
		plt.scatter(lat,lon,s=5,color=col[idx],label=idx)

print('It cost '+str(round(time.time()-t_start,6))+' s.')
plt.title('cluster with dbscan cluster_number='+str(k))
plt.xlabel('latitude')
plt.ylabel('longitude')
plt.legend()
plt.show()
```

結果確實把右上方的outlier排除掉了，但分群的結果似乎不太理想，例如下方兩三顆點的也被分為同一群，應該要再調整eps和min_samples，這次將min_samples=3

但有個嚴重的問題，x和y軸代表的是經緯度，我們應該是要以實際的距離作為clustering的依據，而不是地圖上的距離，因此這邊的metric改用haversine，這可以用great_circle來計算距離，而eps則是km/6371，將之轉變成radians

```python
t_start=time.time()
dbscan=DBSCAN(eps=50/6371,min_samples=3,metric='haversine').fit(data)
plt.figure(figsize=(7,6),dpi=100)
k=len(set(dbscan.labels_))-(1 if -1 in dbscan.labels_ else 0)

for idx in range(-1,k):
	lat=[]
	lon=[]
	for index,value in data.iterrows():
		if(dbscan.labels_[index]==idx):
			lat.append(value['end station latitude'])
			lon.append(value['end station longitude'])
	if idx==-1:
		plt.scatter(lat,lon,s=5,color='black',label='outlier')
	else:
		plt.scatter(lat,lon,s=5,color=col[idx],label=idx)

print('It cost '+str(round(time.time()-t_start,6))+' s.')
plt.title('cluster with dbscan cluster_number='+str(k))
plt.xlabel('latitude')
plt.ylabel('longitude')
plt.legend()
plt.show()
```

若將距離設為50km，可以看到結果很漂亮的分出不同群，而在實際地圖上，左右兩半邊的點被一條河流所切割，所以這樣的分法也是非常合理的。但這樣的分似乎又太兩極化了，如果將eps降低到30km呢？

```python
t_start=time.time()
dbscan=DBSCAN(eps=30/6371,min_samples=3,metric='haversine').fit(data)
plt.figure(figsize=(7,6),dpi=100)
k=len(set(dbscan.labels_))-(1 if -1 in dbscan.labels_ else 0)

for idx in range(-1,k):
	lat=[]
	lon=[]
	for index,value in data.iterrows():
		if(dbscan.labels_[index]==idx):
			lat.append(value['end station latitude'])
			lon.append(value['end station longitude'])
	if idx==-1:
		plt.scatter(lat,lon,s=5,color='black',label='outlier')
	else:
		plt.scatter(lat,lon,s=5,color=col[idx],label=idx)

print('It cost '+str(round(time.time()-t_start,6))+' s.')
plt.title('cluster with dbscan cluster_number='+str(k))
plt.xlabel('latitude')
plt.ylabel('longitude')
plt.legend()
plt.show()
```

可以看到上方又被分割出了一群，是還滿驚喜的結果，但下方的outlier變得更多則是無奈的地方

## Kmeans V.S. DBScan

時間上，Kmeans的時間取決於k的大小，k愈大則要花愈多的時間，時間由0.2~0.5秒不等，而DBScan則是很穩定的在0.19秒上下

分群效果上，Kmeans受限於自身演算法的侷限，outlier的影響可能要靠k設大一點來劃出來，但如果是outlier很多的資料時，這個方法顯然不太實用。相比於DBScan，他對於outlier的抵抗比Kmeans強了許多，但在於eps、min_samples要設定多少，是個難以決定的問題，另外就是出來的結果可能outlier過多，造成分群結果有些侷限

另外因為DBScan是由一個最低距離來決定每個分群，所以在例如上方的例子，緯度在-74.025~-74.075的點很明顯的可以被分為一群，但因為點之間距離太鬆散，所以會被分成outlier。這在處理每個群舒密程度不一時的狀況會有不好的結果。

# Temporal clustering

## Agglomerative Clustering

```python
data=pd.read_csv('./inout_flow_data.csv')
d=[]
ids=[]

for i in range(1,3479):
    now=data.loc[(data['station_id']==i)&(data['time']>=0)&(data['time']<336),'in_flow_count']
    if(now.value_counts()[0]!=336):
        ids.append('in'+str(i))
        d.append(now)
    now=data.loc[(data['station_id']==i)&(data['time']>=0)&(data['time']<336),'out_flow_count']
    if(now.value_counts()[0]!=336):
        ids.append('out'+str(i))
        d.append(now)
```

上面我將各station一週內的in_flow_count與out_flow_count分別包成兩個336維的點，供下面做運算

首先我也先用了elbow method觀察該分成幾群比較適合

```python
errors=[]
cluster_range=range(1,25)

plt.figure(figsize=(7,5))
for k in cluster_range: #record sum of squared distances to their closest center for each k values 
	kmeans=KMeans(n_clusters=k,n_init=15).fit(d)
	errors.append(kmeans.inertia_)
plt.plot(cluster_range,errors,marker='o')
plt.xticks(np.linspace(0,20,21))
plt.title('elbow method')
plt.xlabel('cluster number')
plt.ylabel('sum of squared distances')
plt.show()
```

可以看到雖然斜率轉折較明顯的點大概是k=3，但distance仍然是很高，或許可以再試試看k=14的結果

下面將先用scipy的hierarchy畫圖，列出各群的個數，再詳細畫圖

```python
from scipy.cluster.hierarchy import linkage, fcluster, dendrogram
Z = linkage(d,method='ward')

plt.figure(figsize=(6,4))
dendrogram(Z,p=3,truncate_mode='lastp')
plt.show()
```

```python
k=3
t=np.arange(336)
t_start=time.time()
ac=AgglomerativeClustering(n_clusters=k).fit(d)
plt.figure(figsize=(8,7),dpi=100)
for idx in range(0,k):
    pnow=plt.subplot(k,1,idx+1)
    for i in range(len(d)):
        if(ac.labels_[i]==idx):
            pnow.plot(t,d[i],linewidth=1)
            pnow.set_xlabel('time')
            pnow.set_ylabel('count')
print('It cost '+str(round(time.time()-t_start,6))+' s.')
plt.show()
```

可以看到分成三群已經可以有大致的規律，但效果並不是很好

```python
Z = linkage(d,method='ward')

plt.figure(figsize=(10,6))
dendrogram(Z,p=14,truncate_mode='lastp')
plt.show()
```

```python
k=14
t=np.arange(336)
t_start=time.time()
ac=AgglomerativeClustering(n_clusters=k).fit(d)
plt.figure(figsize=(8,7),dpi=100)
for idx in range(0,k):
    pnow=plt.subplot(k,1,idx+1)
    for i in range(len(d)):
        if(ac.labels_[i]==idx):
            pnow.plot(t,d[i],linewidth=1)
            pnow.set_xlabel('time')
            pnow.set_ylabel('count')
print('It cost '+str(round(time.time()-t_start,6))+' s.')
plt.show()
```

分成14群則是可以擁有不錯的結果，尤其是倒數第3群，兩個車站的流量竟然如此相似

而這邊的affinity則都是預設的euclidean，使用其他種類的affinity都會顯示不被允許

## Use PCA

```python
from sklearn.decomposition import PCA
```

首先先試個比較極端的例子，將7\*48維度變成7\*1維度來看看

```python
dnew=[]
for i in d:
    dnew.append(PCA(n_components=1).fit_transform(i.values.reshape(7,48)).reshape(-1))
    
Z = linkage(dnew,method='ward')
plt.figure(figsize=(16,6))
pnow=plt.subplot(121)
dendrogram(Z,p=3,truncate_mode='lastp')
pnow=plt.subplot(122)
dendrogram(Z,p=14,truncate_mode='lastp')
plt.show()
```

左方分成3群的情況，顯然與原先分成(814,321,98)有明顯的不同，而右方分成14群與上圖相比，不僅每群的個數除了少數幾群相同之外，其他都差了不少，而且分群的路徑也由本來直接分出(464,350)這兩個比較大的分支，變成平均的分散開來。總結的說降低維度太多其實影響滿大的

再來試試看由7\*48維度降為7\*24維度

```python
dnew=[]
for i in d:
    dnew.append(PCA(n_components=24).fit_transform(i.values.reshape(7,48)).reshape(-1))

Z = linkage(dnew,method='ward')
plt.figure(figsize=(16,6))
pnow=plt.subplot(121)
dendrogram(Z,p=3,truncate_mode='lastp')
pnow=plt.subplot(122)
dendrogram(Z,p=14,truncate_mode='lastp')
plt.show()
```

結果依然跟原先的差很多，因此我決定將投影到跟本來維度一樣的空間，再看看結果是不是也不一樣

```python
dnew=[]
for i in d:
    dnew.append(PCA(n_components=48).fit_transform(i.values.reshape(7,48)).reshape(-1))

Z = linkage(dnew,method='ward')
plt.figure(figsize=(16,6))
pnow=plt.subplot(121)
dendrogram(Z,p=3,truncate_mode='lastp')
pnow=plt.subplot(122)
dendrogram(Z,p=14,truncate_mode='lastp')
plt.show()
```

```python
plt.plot(dnew[0])
plt.show()
```

令我驚訝的，分群結果竟然跟投影到7\*24維度是一樣的，而且即便投影的維度跟本來一樣，分群結果依然與原本不同。

實際一個一個測試之後，發現7\*6維度以上的分群結果都會相同，直到7\*5以下的維度才會不一樣，這或許是因為PCA內部的數學原理，已至於可能要到更深的層才會有差別，而這邊取的較淺，所以才會看不出來差別

```python
dnew=[]
for i in d:
    dnew.append(PCA(n_components=6).fit_transform(i.values.reshape(7,48)).reshape(-1))

k=14
t=np.arange(336)
t_start=time.time()
ac=AgglomerativeClustering(n_clusters=k).fit(dnew)
plt.figure(figsize=(8,7),dpi=100)
for idx in range(0,k):
    pnow=plt.subplot(k,1,idx+1)
    for i in range(len(dnew)):
        if(ac.labels_[i]==idx):
            pnow.plot(t,d[i],linewidth=1)
            pnow.set_xlabel('time')
            pnow.set_ylabel('count')
print('It cost '+str(round(time.time()-t_start,6))+' s.')
plt.show()
```

而降到7\*6維之後，速度由3.244397降至3.040169，但分群結果似乎比原先的糟糕很多，或許不太適合用PCA來處理

# Other

我想看看若以流量來分群，畫出來的車站分佈圖長什麼樣子

```python
data=pd.read_csv('./station_data.csv')
data1=data.iloc[:,[2,3]]
data2=pd.read_csv('./inout_flow_data.csv')
d=[]
ids=list(data.loc[:,'end station id'])

for i in ids:
	now=data2.loc[(data2['station_id']==i)&(data2['time']>=0)&(data2['time']<336),'in_flow_count']
	d.append(now)
```

```python
plt.figure(figsize=(12,5),dpi=200)
k=3
ac=AgglomerativeClustering(n_clusters=k).fit(d)
pnow=plt.subplot(121)
pnow.set_title('cluster with AgglomerativeClustering k='+str(k))
pnow.set_xlabel('latitude')
pnow.set_ylabel('longitude')
for idx in range(0,k):
	lat=[]
	lon=[]
	for index,value in data1.iterrows():
		if(ac.labels_[index]==idx):
			lat.append(value['end station latitude'])
			lon.append(value['end station longitude'])
	pnow.scatter(lat,lon,s=5,label=idx)
plt.legend()

k=14
ac=AgglomerativeClustering(n_clusters=k).fit(d)
pnow=plt.subplot(122)
pnow.set_title('cluster with AgglomerativeClustering k='+str(k))
pnow.set_xlabel('latitude')
pnow.set_ylabel('longitude')
for idx in range(0,k):
	lat=[]
	lon=[]
	for index,value in data1.iterrows():
		if(ac.labels_[index]==idx):
			lat.append(value['end station latitude'])
			lon.append(value['end station longitude'])
	pnow.scatter(lat,lon,s=5,label=idx)
plt.legend()

plt.show()
```

可以看到，不管是分成3群還是分成14群，每群大都是一叢一叢分佈，尤其是左圖，圖的左測幾乎都是第0群，而右測中間幾乎都是第1群，或許表示地理位置與生活習慣有些關係
